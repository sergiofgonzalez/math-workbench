# Drawing a 2D dino in Python

> this notebook explains how draw a 2D dino in Python

First thing is to import the `vector_drawing.py` library

```python
from vector_drawing import *
```

Next, we create a list of the points that delimit the dino shape 

```python
dino_points = [(6, 4), (3, 1), (1, 2), (-1, 5), (-2, 5), (-3, 4), (-4, 4), (-5, 3), (-5, 2), (-2, 2), (-5, 1), (-4, 0), (-2, 1), (-1, 0),
                        (0, -3), (-1, -4), (1, -4), (2, -3), (1, -2), (3, -1), (5, 1)]
```

After that, we use `draw(...)` to plot the points of the dino in a chart

```python
draw(Points(*dino_points))
```

Now we need to draw lines between the points in a given order to draw the dino. This can be done using `draw(...)` and passing the corresponding segments as seen below: 

```python
draw(Segment((6, 4), (3, 1)))
```

The function `draw(...)` allows you to pass multiple `Segment`s

```python
draw(Segment((6, 4), (3, 1)), Segment((3, 1), (1, 2)))
```

As hardcoding all the `Segment`s manually one-by-one is very tedious, we define a function that returns all the segments for the dino shape given the points that make up the dino shape

```python
def get_dino_segments(*vectors):
  segments = []
  start = vectors[0]

  for v in vectors[1:-1]:
    segments.append(Segment(start, v))
    start = v

  segments.append(Segment(v, vectors[-1]))
  segments.append(Segment(vectors[-1], vectors[0]))
  return segments
```

Right after that, we can invoke `draw(...)` to create the final dino shape

| NOTE: |
| :---- |
| See how we are invoking `draw()` passing the collection of dino points along with the segments in one single invocation. |

```python
dino_segments = get_dino_segments(*dino_points)
draw(Points(*dino_points), *dino_segments)
```
