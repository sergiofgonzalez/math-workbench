## Adjusting the grid and aspect ratio

Sometimes you will need to adjust the grid and the aspect ratio to obtain nice charts when using `draw(...)`.

Consider the following example that draws the points for a *parabola*.

When using the default options, the resulting graph will look enlarged on the y-axis:

```python
from my_vector_drawing_lib import Points, draw

vectors = [(x, x ** 2) for x in range(-10, 11)]
points = Points(*vectors)

draw(points)
```

You can start by adjusting the `grid` parameter, so that only 1 every 10 units in the y-axis is shown, but that won't fix the size and ratio:

```python
draw(points, grid=(1, 10))
```

Using `nice_aspect_ratio` parameter won't solve the problem on its own either, although now the size of the graph is exactly what we are looking for.

```python
draw(points, nice_aspect_ratio=False)
```

Adjusting both the `grid` and `nice_aspect_ratio` parameters do the trick:

```python
draw(points, grid=(1, 10), nice_aspect_ratio=False)
```
